```python
final_df = US_data.merge(c_data, how = "inner", left_on =  'NAME', right_on ='Province_State')
final_df['case category'] = 0;
final_df = final_df.drop_duplicates(subset=['STATE'])
final_df = final_df.loc[final_df['NAME']!='Guam']
final_df = final_df.loc[final_df['NAME']!='American Samoa']
final_df = final_df.loc[final_df['NAME']!='District of Columbia']
final_df = final_df.loc[final_df['NAME']!='Virgin Islands']
final_df = final_df.reset_index()
```

```python
# 0 to 1000, 1001 to 5000, 5001 to 10000, 100001 to 20000, 20001 to 50,000, 50,0001 to 100000, 100,000 or more
count = 0
for x in list(final_df['Confirmed']):
    if x > 100000:
        final_df.loc[count,'case category'] = 'F: 100,000 or more'
    elif  x > 50000:
        final_df.loc[count,'case category'] = 'E: 50,001 to 100,000'
    elif x > 20000:
        final_df.loc[count,'case category'] = 'D: 20,001 to 50,000'
    elif x > 10000:
        final_df.loc[count,'case category'] = 'C: 10,001 to 20,000'
    elif x > 5000:
        final_df.loc[count,'case category'] = 'B: 5,001 to 10,000'
    elif x > 1000:
        final_df.loc[count,'case category'] = 'A: 1,001 to 5,000'
    else:
        final_df.loc[count,'case category'] = '1: 0 to 1,000'
    count += 1  
final_df.sort_values('Confirmed').head()
```
